# Librerias

```
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.mixture import GaussianMixture
from sklearn.cluster import MeanShift
from sklearn.decomposition import PCA
from sklearn.cluster import Birch
import joblib
```

# Carga de datos

```
file_path = '../data/base_prueba_tecnica.xlsx'
df = pd.read_excel(file_path)
df.head()
```

```
numeric_columns = ['Edad', 'Ingresos Anuales', 'Gasto Mensual', 'Puntuación de Crédito', 'Comportamiento en linea', 'Compras en el ultimo año']


scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[numeric_columns] = scaler.fit_transform(df[numeric_columns])


pca = PCA(n_components=3)
X_reduced = pca.fit_transform(df_scaled[numeric_columns])

x = X_reduced[:, 0]
y = X_reduced[:, 1]
z = X_reduced[:, 2]
```

# Entrenamiento y Persistencia del Modelo BIRCH
Entrenamos un modelo BIRCH con los datos de dimensionalidad reducida y especificamos el número de clusters deseado. 

```
# Entrenar el modelo BIRCH
birch_model = Birch(n_clusters=3) 
birch_model.fit(X_reduced)

# Guardar el modelo BIRCH
joblib.dump(scaler, '../back/scaler.pkl')
joblib.dump(pca, '../back/pca.pkl')
joblib.dump(birch_model, '../back/birch_model.pkl')
```

# Entrenamiento y Persistencia de Modelos Adicionales
Además del modelo BIRCH, entrenamos modelos adicionales de clustering, como Gaussian Mixture y Mean Shift. Cada modelo se ajusta a los datos de dimensionalidad reducida, y luego se guarda en el sistema de archivos para su uso futuro. Esto  permite comparar fácilmente diferentes modelos y seleccionar el más adecuado para nuestras necesidades específicas de clustering.


```
models = [
    ('Gaussian Mixture', GaussianMixture(n_components=3)), 
    ('Mean Shift', MeanShift(bandwidth=None)),
]


for name, model in models:
    model.fit(X_reduced)
    # Guardar el modelo 
    joblib.dump(model, f'../back/{name.replace(" ", "_").lower()}_model.pkl')
```
